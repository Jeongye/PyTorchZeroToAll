```python
import torch
from torch.autograd import Variable


# define our data with metrics
x_data = Variable(torch.Tensor([[1.0], [2.0], [3.0]]))    # 3 by 1 tensor
y_data = Variable(torch.Tensor([[2.0], [4.0], [6.0]]))

class Model(torch.nn.Module):
  def __init__(self):
    """
    In the constructor we instantiate two nn.Linear module
    """
    super(Model, self).__init__()
    # one linear block
    self.linear = torch.nn.Linear(1, 1)    # input size is one, output size is one
  
  def forward(self, x):
    """
    In the forward function we accept a Variable of input data and we must return
    a Variable of output data. We can use Modules defined in the constructor as
    well as arbitrary operators on Variables.
    """
    y_pred = self.linear(x)    # not gonna use x*w! we use the block that we initialized init
    return y_pred

# our model
model = Model()


# Construct our loss function and an Optimizer. The call to model.parameters()
# in the SGD constructor will contain the larnable parameters of the two
# nn,Linear modules which are members of the model
criterion = torch.nn.MSELoss(size_average=False)
# model.parameters() can provide all the variables we need to be updated
optimizer = torch.optim.SGD(model.parameters(), lr=0.01)


# Training loop
for epoch in range(500):
  # Forward pass: Compare predicted y by passing x to the model
  y_pred = model(x_data)
  #print(epoch, y_pred)

  # Compute and print loss
  loss = criterion(y_pred, y_data)
  print(epoch, loss.data)    # 영상에서는 loss.data[0]으로 나와있는데 그렇게 하면 오류가 뜬다... 왜지?

  # Zero gradients, perform a backward pass, and update the weights.
  optimizer.zero_grad()
  loss.backward()
  optimizer.step()    # to update variable


# After training
hour_var = Variable(torch.Tensor([[3.0]]))
print("predict (after training)", 3, model.forward(hour_var).data[0][0])
```
