```python
import torch
import torch.nn.functional as F
from torch.autograd import Variable


# define our data with metrics
x_data = Variable(torch.Tensor([[1.0], [2.0], [3.0], [4.0]]))
y_data = Variable(torch.Tensor([[0.], [0.], [1.], [1.]]))

class Model(torch.nn.Module):
  def __init__(self):
    super(Model, self).__init__()
    self.linear = torch.nn.Linear(1, 1)    # input size is one, output size is one
  
  def forward(self, x):
    y_pred = F.sigmoid(self.linear(x))
    return y_pred

# our model
model = Model()


criterion = torch.nn.BCELoss(size_average=False)    # binary cross entrophy
optimizer = torch.optim.SGD(model.parameters(), lr=0.01)


# Training loop
for epoch in range(1000):
  y_pred = model(x_data)
  #print(epoch, y_pred)

  # Compute and print loss
  loss = criterion(y_pred, y_data)
  print(epoch, loss.data.item())    # item()!!!!

  # Zero gradients, perform a backward pass, and update the weights.
  optimizer.zero_grad()
  loss.backward()
  optimizer.step()    # to update variable


# After training
hour_var = Variable(torch.Tensor([[1.0]]))
print("predict 1 hour", 1.0, model(hour_var).data.item() > 0.5)

hour_var = Variable(torch.Tensor([[7.0]]))
print("predict 7 hour", 7.0, model(hour_var).data.item() > 0.5)
```
